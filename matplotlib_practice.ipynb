```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
 # Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.show()
```

```python
#5.1.5 Bar Charts-MATLAB Approach
# Set the x-axis to a list of strings for each month.
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.bar(x_axis,y_axis,color="green",label="Boston")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.show
```

```python
#create Horizontal plot

plt.barh(x_axis, y_axis,color="magenta",label="Boston")
#inrevert y_axis
plt.gca().invert_yaxis()
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
```

```python
#5.17 Create a Scatter Plot Using the MATLAB Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]


```

```python
plt.plot(x_axis,y_axis,"o")
```

```python
#method two
plt.scatter(y_axis,x_axis,c="Red",label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()

```

```python
#Create a Bubble Chart Using the MATLAB Method
#cont.
plt.scatter(x_axis, y_axis,s=y_axis)
```

```python
#change the size using a for loop to create the list y_axis_larger
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
    

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
##change the size using List Comprehension

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Create a Scatter Plot Using the Object-Oriented Interface
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Create a Bubble Chart Using the Object-Oriented Interface
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],c="skyblue",alpha=0.8, label="Boston",linewidths=2,edgecolors="black")
plt.gca().invert_yaxis()

ax.set_xlim(0, 50)
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.legend()
```
